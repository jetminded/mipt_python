Сделаем функцию, которая генерирует обёртку-таймер для любой переданной ей функции

```python
import time

def measure_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Function '{func.__name__}' run {execution_time:.4f} seconds")
        return result
    return wrapper


def slow_function(number: int):
    '''Very slow function'''
    time.sleep(2)
    return f"Done {number}"

result = measure_time(slow_function)(10)
print(result)
```

И `measure_time` уже декоратор. Добавим его через `@` перед определением функции

```python
@measure_time
def slow_function(number: int):
    '''Very slow function'''
    time.sleep(2)
    return f"Done {number}"


result = slow_function(10)
print(result)
```

Проблема, у обёрнутой функции поменялись метаданные на `wrapper`. Но они нужны для подсказок и stacktrace (в случае ошибки)

```python
slow_function.__name__, slow_function.__doc__
```

Сделаем автоматическое восстановление через добавление декоратора `wraps` в генерируемую обёртку

```python
import time
from functools import wraps

def measure_time(func):
    @wraps(func)  # используем этот декоратор, чтобы восстановить метаданные
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Function '{func.__name__}' run {execution_time:.4f} seconds")
        return result
    return wrapper


@measure_time
def slow_function(number: int):
    '''Very slow function'''
    time.sleep(2)
    return f"Done {number}"

result = slow_function(10)
print(result)
```

```python
slow_function.__name__, slow_function.__doc__
```

Мой любимый декоратор: `lru_cache` - кэширует аутпут функции по инпуту (аргументам)

```python
from functools import lru_cache

@lru_cache
def fib(num: int):
    # специально неэффективно реализуем
    if num <= 1:
        return 1
    else:
        return fib(num - 1) + fib(num - 2)
    
fib(1000)
```
