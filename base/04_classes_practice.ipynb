## 1. Абстрактный класс

Создайте абстрактный класс `LibraryItem`.

### Требования:

1. Используйте модуль `abc` (а какие есть другие варианты?).
2. Конструктор должен принимать:
   - `title`
   - `year`
3. Определите:
   - абстрактный метод `get_description()`
   - абстрактный метод `calculate_late_fee(days_late)`
4. Добавьте свойство (property) `age`, которое вычисляет возраст объекта на основе текущего года.
5. Переопределите метод `__str__()` для удобного строкового представления объекта.


````python
from abc import ABC, abstractmethod
from datetime import datetime

class LibraryItem(Borrowable, ABC):
    def __init__(self, title: str, year: int, *args, **kwargs):
        super().__init__()
        self.title = title
        self.year = year
    
    @abstractmethod
    def get_description(self) -> str:
        pass

    @abstractmethod
    def calculate_late_fee(self, days_late) -> float:
        pass

    @property
    def age(self) -> int:
        return datetime.now().year - self.year
    
    def __str__(self):
        # return f'title: {self.title}, year: {self.year}'
        return self.get_description()
    
    def __repr__(self):
        return self.__str__()
````

## `AncientTome`

Дополнительные атрибуты:
- `origin_century`
- `magic_level` - значение из Enum'a MAGIC_LEVEL

Правило расчёта штрафа:
- `2 * days_late * magic_level`

Требуется:
- Переопределить `get_description()`
- Реализовать `calculate_late_fee(days_late)`

````python
from enum import Enum


class MAGIC_LEVEL(Enum):
    COMMON = 1
    RARE = 2
    EPIC = 3
    LEGENDARY = 4
    CHAMPION = 5


class AncientTome(LibraryItem):
    def __init__(self, title, year, origin_century, magic_level: MAGIC_LEVEL):
        self.origin_century = origin_century
        self.magic_level = magic_level
        super().__init__(title=title, year=year)

    def get_description(self):
        return f'title: {self.title}, year: {self.year}, origin_century: {self.origin_century}, magic_level: {self.magic_level}'

    def calculate_late_fee(self, days_late):
        return 2 * days_late * self.magic_level.value
    

AncientTome(
    'lol',
    1999,
    20,
    magic_level=MAGIC_LEVEL.COMMON
)
````

## 3. Миксин Borrowable

Создайте класс-миксин `Borrowable`.

Требования:

Атрибуты:
- `_is_borrowed`
- `_borrower`

Методы:
- `borrow(user_name)`
- `return_item()`

Свойство:
- `is_borrowed`

!!! Класс `Borrowable` не должен наследоваться от `LibraryItem`.

Теперь измените классы `Book` и `AncientTome`, чтобы они наследовались так:

```python
class Book(Borrowable, LibraryItem):
```

Убедитесь, что конструкторы корректно работают с `super()` при множественном наследовании.


ТУТ МЫ РЕШИЛИ ДОБАВИТЬ `Borrowable` СРАЗУ В `LibraryItem`, ибо оба `Book` и `AncientTome` наследуются от `LibraryItem`

````python
class Borrowable:
    def __init__(self, *args, **kwargs):
        self._is_borrowed: bool = False
        self._borrower: str = None
        super().__init__(*args, **kwargs)

    def borrow(self, user_name: str) -> None:
        self._is_borrowed = True
        self._borrower = user_name
    
    def return_item(self, user_name: str) -> None:
        self._is_borrowed = False
        self._borrower = None

    @property
    def is_borrowed(self):
        return self._is_borrowed
````

## 4. Class Methods и Static Methods

Расширьте класс `Book`.

Добавьте атрибут класса:
- `total_books_created`

Добавьте classmethod:
- `from_string(cls, data_string)`
  - Формат строки: `"title,author,year,pages"`
  - Метод должен возвращать объект `Book`

Добавьте staticmethod:
- `is_valid_year(year)`
  - Возвращает `True`, если год находится в диапазоне от 0 до текущего года.

Требования:
- `__init__` должен увеличивать `total_books_created` при создании каждого экземпляра (и уменьшаться при удалении).

````python
class Book(LibraryItem):
    total_books_created = 0

    def __init__(self, title, year, author, pages):
        self.author = author
        self.pages = pages
        Book.total_books_created += 1
        super().__init__(title=title, year=year)

    def __del__(self):
        Book.total_books_created -= 1

    @classmethod
    def from_string(cls, data_str: str):
        return cls(*data_str.strip('"').split(','))

    def get_description(self):
        return f'title: {self.title}, year: {self.year}, author: {self.author}, pages: {self.pages}'
    
    def calculate_late_fee(self, days_late):
        return 0.5 * days_late
    
    @staticmethod
    def is_valid_year(year):
        return 0 < year < datetime.now().year
    
book = Book(
    title='сборник по матану 19-го года',
    year=2019,
    author='Воронцов',
    pages=1024
)

book_2 = Book(
    title='сборник по матану 19-го года',
    year=2019,
    author='Воронцов',
    pages=1024
)

book
````

# 5. Library

Создайте класс `Library`.

Атрибуты:
- `name`
- `collection` (список объектов)

Какие из этих атрибутов надо скрывать от пользователей, чтобы была корректная инкапсуляция? Добавь нижние подчеркивания (одно или два) к тем атрибутам, которым это нужно.

Методы:
- `add_item(item)`
- `remove_item(item)`
- `list_available_items()` — возвращает только те объекты, которые не заимствованы
- `find_by_title(title)`

Class Method:
- `create_default_library(cls)`
  - Создаёт библиотеку, содержащую минимум:
    - один `Book`
    - один `AncientTome`

````python
class Library:
    def __init__(self, name, collection):
        self.name = name
        self.__collection: list[LibraryItem] = collection
    
    def add_item(self, item: LibraryItem):
        self.__collection.append(item)

    def remove_item(self, item: LibraryItem):
        self.__collection.pop(item)

    def list_available_items(self):
        borrowable = []
        for item in self.__collection:
            if item.is_borrowed:
                borrowable.append(item)
        return borrowable

    def find_by_title(self, title):
        borrowable = []
        for item in self.__collection:
            if item.title == title:
                borrowable.append(item)
        return borrowable

    @classmethod
    def create_default_library(cls):
        return cls(
            'my library',
            [
                Book(
                    title='сборник по матану 19-го года',
                    year=2019,
                    author='Воронцов',
                    pages=1024
                )
            ]
        )
````
