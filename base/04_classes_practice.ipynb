## Задача 1.

````python
# Найди и исправь ошибку в инициализаторе класса

class Book:
    def __init__(self, title, author):
        title = title
        author = author

my_book = Book('1984', 'George Orwell')
print(my_book.title, my_book.author)

# Ожидаемый результат: 1984 George Orwell
````

# Задача 2.

````python
# Исправь ошибку в методе класса

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def description():
        return 'Book: ' + title + ', author: ' + author

my_book = Book('The Hobbit', 'J.R.R. Tolkien')
print(my_book.description())

# Ожидаемый результат: Book: The Hobbit, author: J.R.R. Tolkien
````

## Задача 3.

````python
# Исправь ошибку в вызове метода

class Spellbook:
    def __init__(self, name):
        self.name = name

    def cast_spell(self):
        return self.name + ' glows with magical energy!'

spellbook = Spellbook('The Book of Shadows')
print(Spellbook.cast_spell())

# Expected output: The Book of Shadows glows with magical energy!
````
